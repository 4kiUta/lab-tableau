To see if apps are bigger in size they are more expesive

```python
import pandas as pd
import re
```

```python
app = pd.read_csv('AppleStore.csv')
```

```python
app.head()
```

```python
app.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
app.isna().sum()
```

```python
payd_apps = app.loc[(app['price'] > 0)]
```

```python
payd_apps.head()
```

```python
payd_apps.columns
```

```python
payd_apps = payd_apps[['id', 'track_name', 'size_bytes','ver', 'prime_genre','ipadSc_urls.num','price']]
```

```python
payd_apps.dtypes
```

```python
## Clean version from letters

payd_apps['ver'] = payd_apps['ver'].apply(lambda x: re.sub('[A-Za-z]', '', x))
payd_apps['ver'].unique()
```

```python
payd_apps['ver'] = payd_apps['ver'].apply(lambda x: x[0] if len(x) <= 6 else x[:2] if x[0] not in '9876543' else  '0')
```

```python
payd_apps['ver'].unique()
```

```python
payd_apps['ver'] = payd_apps['ver'].apply(lambda x: x.strip('.'))
```

```python
payd_apps['ver']
```

```python
payd_apps['ver'].unique()
```

```python
payd_apps = payd_apps.loc[payd_apps['ver'] != ' ']
```

```python
payd_apps
```

```python
payd_apps.to_csv('App_cleannn.csv')
```
